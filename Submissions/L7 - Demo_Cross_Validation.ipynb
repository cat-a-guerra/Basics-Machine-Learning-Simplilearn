```python
# Import dataset
from sklearn.datasets import load_iris
```

```python
# Load dataset as an array
iris_data = load_iris()
print(iris_data)
```

```python
# Extract the input data and the target values in separate variables
data_input = iris_data.data
data_output = iris_data.target

# Check the input data
print(data_input)
```

```python
# Check the output data
print(data_output)
```

```python
# Split the data into 5 consecutive folds without shuffling by default using K-Fold Cross-Validation 
from sklearn.model_selection import KFold
kf = KFold(n_splits=5, shuffle=True)                  
```

```python
# Iterate the individual train/test sets within the input data using a for loop
for train_set, test_set in kf.split(data_input):
  print("Train set")
  print(train_set)
  print("Test set")
  print(test_set)
```

```python
# Initialize Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rf_class = RandomForestClassifier(n_estimators=10)           # Split data into 10 consecutive folds   

# Import Cross-Validation score library
from sklearn.model_selection import cross_val_score

# Get model evaluated for accuracy using the cross-validation score parameter
print(cross_val_score(rf_class, data_input, data_output, scoring='accuracy', cv=10))
```

```python
# Get accuracy
accuracy = cross_val_score(rf_class, data_input, data_output, scoring='accuracy', cv=10).mean()*100
print('Accuracy of Random Forests: ', accuracy)
```
